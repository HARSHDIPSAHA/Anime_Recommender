```python
data=pd.read_csv("/kaggle/input/myanimelist-dataset/anime-dataset-2023.csv")
rate=pd.read_csv("/kaggle/input/myanimelist-dataset/users-score-2023.csv",nrows=100000000)
```

```python
ratings_with_name = rate.merge(data,on='anime_id')
```

```python
ratings_with_name.head()
```

```python
x = ratings_with_name.groupby('user_id').count()['rating'] > 200
bade_log_users = x[x].index
#taking those users who have voted more than 200 times
```

```python
filtered_rating = ratings_with_name[ratings_with_name['user_id'].isin(bade_log_users)]
```

```python
y = filtered_rating.groupby('English name').count()['rating']>=50
famous_anime = y[y].index
#taking anime which are voted more than 50
```

```python
final_ratings = filtered_rating[filtered_rating['English name'].isin(famous_anime)]
```

```python
table = final_ratings.pivot_table(index='English name',columns='user_id',values='rating')
```

```python
table
```

```python
table.fillna(0,inplace=True)
table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(table)
```

```python
def recommend(anime):
    # index fetch
    index = np.where(table.index==anime)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    l = []
    for i in similar_items:
        item = []
        temp_df = data[data['English name'] == table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('English name')['English name'].values))
        item.extend(list(temp_df.drop_duplicates('English name')['Synopsis'].values))
        item.extend(list(temp_df.drop_duplicates('English name')['Genres'].values))
        
        l.append(item)
    
    return l
```

```python
recommend("Attack on Titan")
```
